# 1. Basic Principles of Planet Detection

Notation with subscripts $∗$ and $p$ refer to quantities for the star and planet, respectively.

Therefore, a `star` has 
- mass $M_∗$ 
- radius $R_∗$ 
- mean density $ρ_∗$ 
- surface gravity $g_∗$ 
- effective temperature $T_∗$

It's `orbiting planet` has the following properties 
- mass $M_p$ 
- radius $R_p$ 
- density $ρ_p$ 
- temperature $T_p$
- surface gravity $g_p$ 

The `orbit` has a 
- semimajor axis $a$ 
- period $P$ 
- eccentricity $e$

## 1.1 Spectroscopic Binaries and Orbital Elements

- Conservation of momentum requires that as a planet orbits a distant star, the star executes a smaller, opposite orbit about their common center of mass. 
  - The star's gravity affects the planet's orbit and vice versa
- The size (and velocity) of this orbit is smaller than that of the planet by a factor of the ratio of their masses. 
- The component of this motion along the line of sight to the Earth can, in principle, be detected as a variable radial velocity. 
- The mass of the exoplanet can be calculated from the magnitude of the radial velocity (RV)

Two mutually orbiting bodies revolve in ellipses about a common center of mass, the origin of our coordinate system

The physical size of the ellipse, given by the semi-major axis, $a$, is set by Newton’s modification of Kepler’s third law of planetary motion:

\begin{align}
P^2= \frac{4*π^2}{G(M_∗+ M_p)} *a^3 
\end{align}

The semimajor axis is $a=a_∗+a_p$, where $a_∗$ and $a_p$ are the `semi-major axes` of the two bodies’ orbits with respect to the center of mass, given by

\begin{align}
a_∗ = \frac{M_p} {M_∗ + M_p} a
\end{align}

\begin{align}
a_p = \frac{M_∗} {M_∗ + M_p} a 
\end{align}

The position of either body in its orbit about the origin can be expressed in `polar coor-dinates (r, ν)`, where $ν$ is the `true anomaly`, the angle between the location of the object and the periastron. 

For elliptic orbits, the true anomaly ν can be calculated from orbital state vectors as:

\begin{align} 
ν = arccos(\frac{e ⋅ r}{|e||r|})
\end{align}

(if $r ⋅ v < 0$ then replace $ν$ by $2π − ν$)

where:

- v = the orbital velocity vector of the orbiting body,
- e = the eccentricity vector,
- r is the orbital position vector

The separation between the star and planet is given by 

\begin{align}
r*(1 +e*cos(ν)) =a*(1−e^2) 
\end{align}

- $e$ is called the `eccentricity of the orbit`, and has the domain [0,1) for bound orbits. 


grafik.png

Practical computation of a body’s position in its orbit with time is usually performed through the intermediate variable 

$E$ = `eccentric anomaly`. 

$E$ is related to the times once periastron passage $T_0$ through the mean anomaly M:


\begin{align}
M = \frac{2π(t−T_0)}{P} = E − e*sin(E)
\end{align}

It allows the computation of $ν$ = true anomaly measured with respect to periastron, through the relation:
\begin{align}
tan \frac{ν}{2}=\sqrt{\frac{1 +e}{1−e} * tan\frac{E}{2}}
\end{align}

The eccentric anomaly is also useful because it is simply related to $r$:
\begin{align}
E= arccos\frac{1−r/a}{e}
\end{align}

### 1.2.1 Doppler measurements

The radial reflex motion of a star in response to an orbiting planet can be measured through precise Doppler measurement, and this motion reveals:
- $P$ = the period 
- distance `(from the observer / planet Earth?)`
- and shape of the orbit `(eccentricity and semimajor axis?)`

### 1.2.2 Parameters

Six parameters determine the **functional form of the periodic radial velocity variations,$V_r$** and thus the observables in a spectroscopic orbit of the star:

Responsible for RV curve:
- $P$ = Period,
- $K$ = Amplitude of RV curve / the semi-amplitude of the signal in units of velocity,
- $T_0$ = Phase ,  

Shape of the radial velocity signature of an orbiting companion:

- $e$ = Eccentricity of the planet's orbit,
- $ω_∗$ = The longitude of the periastron* of the star, 
- `$γ$ =  The bulk velocity of the center of mass of the system` (?)

Two additional orbital parameters complete the description of a planet’s orbit. **These parameters cannot be determined with radial velocity observations alone, and can only be measured through astrometry**: 

- $i$ = inclination of the orbit  
- $Ω$ = longitude of the ascending node

`The apsides refer to the farthest (1) and nearest (2) points reached by an orbiting planetary body (1 and 2) with respect to a primary, or host, body.When discussing orbits around other stars the terms become apastron(1) and periastron(2).`

For example, the apsides of Earth's orbit of the Sun are two: the apsis for Earth's farthest point from the Sun, dubbed the aphelion (1); and the apsis for Earth's nearest point, the perihelion (2)

### 1.2.4 Additional orbital parameters $i$ and $Ω$

- The inclination of the orbit, $i$, determines the angle between the orbital plane and the plane of the sky
- $i= 0$ corresponds to a face-on, counter-clockwise orbit
- $Ω$ is the longitude of the ascending node. 

The effect of the inclination of the orbit is to reduce the radial component of the velocity of the star by $sin(i)$. The fundamental observable of a spectroscopic binary which constrains the physical properties of system is thus 

\begin{align}
\frac{P*K^3(1−e2)^\frac{3}{2}}{2πG}= \frac{M^3_p * sin^3(i)} {(M_p+M_∗)^2}
\end{align}


- $G$ = Newton’s gravitational constant. 
- The right hand side of this equation is known as **the mass function of the system**. 
- For exoplanets where $M_∗$ can be estimated from stellar models, the minimum value for M_p is called **the “minimum mass” of the plane** 
- It is usually labeled *$M_p * sin(i)$” for succinctness *
- since when $M_p≪M_∗$ its small correction to the denominator is negligible, though not ignored. **(?)**
- The true mass of the detected exoplanet is higher by a factorof $1/sin(i)$, which has a typical (median) value of 1.15 for randomly oriented orbits

## 1.3 The Transit method

The presence of a planetary companion to a star gives rise to a multitude of physical phenomena that manifest themselves via **temporal variations of the flux of the system relative to that of an otherwise identical isolated star**. 
- Typically the largest of these occurs if a fortunate alignment allows a **planet to transit (pass in front of) its host star** from our perspective.
- In this case, the star will exhibit **brief, periodic dimmings** which signal the presence of the planet. 
- The condition for a transit is roughly that the projected separation between the planetand host star at the time of inferior conjunction of the planet is less than the sum of the radii of planet and star, 
  - $r(t_c) * cos(i) ≤ R_∗ + R_p$ 
  - where $r(t_c)$ is the separation of the planet from the host star at
  conjunction and 
  - $R_∗$ and $R_p$ are the radii of the star and planet.
  
Given the definition of $ω_∗$
- $r(t_c) = a(1−e^2)/(1 +e * sin(ω_∗))$ 
- transits occur when the impact parameter of the planet’s orbit with respect to the star in units of the host radius, 
  - $b ≡ \frac{a*cos(i)} {R} * \frac{1−e^2}{1 +e * sin(ω_∗)}$ 
- is less than the sum of the (normalized) radii 
  - $b ≤ 1 +k$ , 
  - where $k≡R_p/R_∗$ Integrating over $i$ assuming isotropic orbits and
  thus a uniform distribution of $cos(i)$ yields the transit probability,
  $P_tr ≡ \frac{R_∗ + R_p}{a} * \frac{1 +e*sin(ω_∗)}{1−e^2}$

For a circular orbit and assuming $k≪1$ , this reduces to the simple expression $Ptr=R∗/a $
