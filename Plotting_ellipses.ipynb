The position of either body in its orbit about the origin can be expressed in `polar coor-dinates (r, ν)`, where $ν$ is the `true anomaly`, the angle between the location of the object and the periastron. 

For elliptic orbits, the true anomaly ν can be calculated from orbital state vectors as:

\begin{align} 
ν = arccos(\frac{e ⋅ r}{|e||r|})
\end{align}

(if $r ⋅ v < 0$ then replace $ν$ by $2π − ν$)

where:

- v = the orbital velocity vector of the orbiting body,
- e = the eccentricity vector,
- r is the orbital position vector

The separation between the star and planet is given by 

\begin{align}
r*(1 +e*cos(ν)) =a*(1−e^2) 
\end{align}

- $e$ is called the `eccentricity of the orbit`, and has the domain [0,1) for bound orbits.

```
from matplotlib.patches import Ellipse

n = 3 # scaling factor 
semi_major = 1 *n # Astronomical Unit
semi_minor = 0.99986* n#

def plot_ellipse(semi_major, semi_minor_list):
  fig, ax = plt.subplots(subplot_kw={'aspect': 'equal'})

  
  for semi_minor in semi_minor_list:
    ellipse = Ellipse((5,5), semi_major, semi_minor, fill = False)

    fig.set_size_inches(8, 8)
    
    ax.add_artist(ellipse)
    ellipse.set_clip_box(ax.bbox)
    ellipse.set_alpha(np.random.rand())
    ellipse.set_facecolor(np.random.rand(3))

    ax.set_xlim(0, 10)
    ax.set_ylim(0, 10)

    plt.show()
```

```
eccentricity = [0, 0.3, 0.6, 0.9]
semi_major_list = [3]

def semi_minor(semi_major, eccentricity):
  
  semi_minor_l = []

  for e in eccentricity:
    for s in semi_major:
      semi_minor_val = s * np.sqrt(1 - e**2)
      semi_minor_l.append(semi_minor_val)

  return semi_minor_l

semi_minor_list = semi_minor(semi_major, eccentricity)
```

```
plot_ellipse(semi_major, semi_minor_list)
```
